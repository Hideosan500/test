```python
import matplotlib as mplt
from matplotlib import pyplot as plt
import numpy as np
plt.style.use('default')

fig0,ax0 = plt.subplots(1)
fig0.set_size_inches(8,4.5)

phaseshift = 1.3

def fx1(x): return 70 * np.sin(x-phaseshift) + 96
def fx2(x): return -70 * np.sin(x-phaseshift) + 96

def fy1(x): return 70 * np.sin(x) + 78
def fy2(x): return -70 * np.sin(x) + 78

x = np.linspace(0,10,100)#[:60]
ax0.plot(x, fx1(x),color = 'C0',linestyle = '-',label='X1')
ax0.plot(x, fx2(x),color = 'C0',linestyle = '--',label='X2')

x = np.linspace(0,10,100)#[40:]
ax0.plot(x, fy1(x),color = 'C1',alpha = 0.4,linestyle = '-',label='Y1')
ax0.plot(x, fy2(x),color = 'C1',alpha = 0.4, linestyle = '--',label='Y2')


ax0.axhline(96, color = 'C0', linestyle = '--')
ax0.axhline(78, color = 'C1', linestyle = '--', alpha = 0.4)
ax0.axhline(0,color='black',linewidth = 0.5)


#ax0.axvline(np.pi/2 + 1.3,color = 'C0',linestyle = '--')
#ax0.axvline(np.pi/2,color = 'C1',linestyle = '--')

ax0.set(xlim = (0,15),ylim =(0,200),xlabel = 'Time',ylabel = 'Voltage')

#noteable points

rightmax_x1 = (np.pi*5.0/2.0+phaseshift,fx1(np.pi*5.0/2.0+phaseshift))
rightmax_x2 = (np.pi*5.0/2.0+phaseshift,fx2(np.pi*5.0/2.0+phaseshift))

leftmax_x1 = (np.pi/2.0+phaseshift,fx1(np.pi/2.0+phaseshift))
leftmax_y1 = (np.pi/2.0,fy1(np.pi/2.0))


#Amplitude Annotation

ax0.annotate('',xy=rightmax_x1, xycoords = 'data',xytext = (rightmax_x1[0]+2,rightmax_x1[1]),
             textcoords = 'data',arrowprops = {'arrowstyle':'-',})
ax0.annotate('Amplitude', xy=(rightmax_x1[0]+2.1,130), xycoords='data', ha='left',
             xytext=(0, 0), textcoords='offset points')
ax0.annotate('', xy=(rightmax_x1[0]+2,rightmax_x1[1]), xytext=(rightmax_x2[0]+2,96),
             xycoords='data',textcoords='data',arrowprops={'arrowstyle': '<->',})

#Offset Annotation
ax0.annotate('', xy=(rightmax_x1[0]+2,0), xytext=(rightmax_x2[0]+2,96),
             xycoords='data',textcoords='data',arrowprops={'arrowstyle': '<->',})

ax0.annotate('Offset',xy=(rightmax_x1[0]+2.1,50), xycoords='data', ha='left',
             xytext=(0, 0), textcoords='offset points')

#Phaseshift Annotation

ax0.annotate('',xy=leftmax_x1, xycoords = 'data',xytext = (leftmax_x1[0],175),
             textcoords = 'data',arrowprops = {'arrowstyle':'-',})

ax0.annotate('',xy=leftmax_y1, xycoords = 'data',xytext = (leftmax_y1[0],175),
             textcoords = 'data',arrowprops = {'arrowstyle':'-',})

ax0.annotate('', xy=(leftmax_y1[0],175), xytext=(leftmax_x1[0],175),
             xycoords='data',textcoords='data',arrowprops={'arrowstyle': '<->',})

ax0.annotate('Phase Shift',xy=(leftmax_y1[0],175), xycoords='data', ha='left',
             xytext=(-10,5), textcoords='offset points')

#Frequency Annotation 

ax0.annotate('',xy=rightmax_x1, xycoords = 'data',xytext = (rightmax_x1[0],175),
             textcoords = 'data',arrowprops = {'arrowstyle':'-',})

ax0.annotate('', xy=(leftmax_x1[0],175), xytext = (rightmax_x1[0],175),
             xycoords='data',textcoords='data',arrowprops={'arrowstyle': '<->',})

ax0.annotate('1/Frequency',xy=(leftmax_x1[0],175), xycoords='data', ha='left',
             xytext=(40,5), textcoords='offset points')

#Romove Tick - Labels 

ax0.xaxis.set_major_formatter(plt.NullFormatter())
ax0.yaxis.set_major_formatter(plt.NullFormatter())

# Add Legend 

plt.legend()

# Save Figure 

fig0.savefig('VoltageAspects.pdf')

```
